# Vamos a comenzar con Python

Una variable en Python es un nombre que se utiliza para referirse a un espacio de memoria en el ordenador, donde puedes almacenar datos. Al asignar un valor a una variable, estás guardando ese valor en el espacio de memoria asociado con ese nombre. Las variables en Python pueden contener diferentes tipos de datos, como números, cadenas de texto, listas, diccionarios y más.

# importar librerias

```python
# vamos aimportar las librerias más comunes
# para ejecutar una celda puedes usar contrl + enter
# shift + enter

import pandas as pd     # sirve para trabajar con dataframes (tablas)
import matplotlib.pyplot as plt   # sirve para realizar plots
import numpy as np    # sirve para generar los datos
```

```python
# Creando una tabla de 50 observaciones con 2 columnas numéricas
# np.random.rand(filas, columnas)

datos = np.random.rand(50, 3)  # genera los datos con 3 columnas
tabla = pd.DataFrame(datos, columns=['Columna 1', 'Columna 2','Columna 3'])
```

```python
tabla.head()  # Mostrando las primeras 5 observaciones para verificar la tabla creada
```

```python
# para verificar que la tabla tiene 50 obs

len(tabla)
```

# ploteando columna 1 con columna 2


```python
# plt.plot(x,y)

x = tabla['Columna 1']
y = tabla['Columna 2']

plt.scatter(x,y, s = 100, color ='g', alpha = 0.5)
plt.title('Diagrama de disperción entre X e y')
plt.xlabel('Columna 1')
plt.ylabel('Columna 2')
plt.grid()
plt.show()
```
